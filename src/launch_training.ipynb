```python
from datagen import *
from clustering import *
from utils import *
from train import *

import torch
import matplotlib.pyplot as plt
```

```python
for dataagrs in [{'imgpath':'../maps/map1.png', 'start':(2064,2114,21,71), 'goal':(21,71,3825,3875),'poolsize':10, 'test_size':0.3} ]:
    dataargs = objectview(dataagrs)

for args in [{'model_type': 'GraphSage', 'dataset': 'maze', 'num_layers': 2, 'heads': 1, 'node_feat_dim':2, 'num_clusters':4, 'batch_size': 32,
              'hidden_dim': 32, 'dropout': 0.5, 'epochs': 500, 'lossft': kmeansloss,'opt': 'adam', 'opt_scheduler': 'none', 'opt_restart': 0,
              'weight_decay': 5e-3, 'lr': 0.001, 'visualize': True,'test':True}]:
    args = objectview(args)

print("Create the maze dataset")
dataset = create_dataset(dataargs)

test_loss, losses, best_model, best_loss, test_loader = train(dataset, args)
print("Minimum loss: {0}".format(min(losses)))

if args.test:
    test(test_loader, best_model, is_validation=False, save_model_preds=False)
    print()

plt.plot(losses, label="training loss" + " - " + args.model_type)
if args.test:
    plt.plot(test_loss, label="test loss" + " - " + args.model_type)
plt.legend()
plt.show()

for batch in test_loader:
    with torch.no_grad():
        pred_softmax, pred_embeddings = best_model(batch)
    mask = batch.test_mask
    pred_test = pred_softmax[mask]

visualize(pred_softmax, pred_embeddings, batch, mask, overlay=True)
```
